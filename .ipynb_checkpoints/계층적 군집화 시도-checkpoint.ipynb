```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score , cross_validate
from sklearn.model_selection import GridSearchCV

pd.options.display.max_columns = 999
pd.options.display.max_rows = 999
```

```python
#df = pd.read_csv('공모전데이터_분석용_KOTRA_0525.csv')
df = pd.read_csv('data_merged.csv',encoding= 'euc-kr')
df.drop(columns = ['UNC_YEAR','COUNTRYKR','impct2Datatype','cmppwBaltrDatatype','year5ImprtGrwrtDatatype',
                   'attrcGdpcpDatatype','cntCdDatatype','attrcGdpDatatype','updDtDatatype',
                  'xportYear2ImprtGrwrtDatatype','impct6Datatype','regDt','year1ImprtGrwrtDatatype',
                  'hsCdDatatype','updDt','impct4Datatype','year2ImprtGrwrtDatatype',
                  'attrcImprtAmtRateDatatype','rankDatatype','acbltAvgTrfrtDatatype',
                  'regDtDatatype','impct1Datatype','mrktAcbltDatatype','mrktCmppwDatatype',
                  'xportAmtDatatype','attrcMrktbGnrlzDatatype','imprtAmtDatatype','bhrcNatnDatatype',
                  'korXportAmtDatatype','mrktGrwthDatatype','grwthMrktbGnrlzDatatype',
                  'relCmdltCoDatatype','impct7Datatype','critYearDatatype','mrktAttrcDatatype',
                  'acbltMrktbGnrlzDatatype','cmppwMrktbGnrlzDatatype','impct5Datatype','impct3Datatype',
                  'attrcImprtAmtDatatype','critYear','parent'], inplace=True)
first_column = df.pop('TOP_HSCD')
df.insert(0, 'TOP_HSCD', first_column)
```

```python
df_new = df.groupby(df['TOP_HSCD']).mean()
df_new = df_new.drop(columns = ['HSCD','COUNTRYCD','NY_GDP_MKTP_CD_1Y','rank',])
df_new = df_new.drop([99])
df_new.shape
```

# 클러스터링 

```python
df_new.columns
```

## 정규화

```python
from sklearn.preprocessing import normalize
df_norm = normalize(df_new)
df_norm = pd.DataFrame(df_norm, columns=df_new.columns)
df_norm.head()
```

## Dendrogram 확인

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title(" Dendograms")
dend = shc.dendrogram(shc.linkage(df_norm, method='ward'))
```

## Agglomerative 알고리즘 사용
* 개별 데이터에서 시작해서 클러스터를 점차 만들어가는 방식

```python
#1. 유클리디안
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
predict = cluster.fit_predict(df_norm[['mrktAcblt','mrktAttrc','mrktGrwth','attrcGdp']])
predict


# ,
```

```python
import collections, numpy

collections.Counter(predict)
```

```python
df_new['predict'] = predict

cols = list(df_new.columns)
cols = [cols[-1]] + cols[:-1]
df_new = df_new[cols]

df_new
```

```python
#2. 코사인 유사도 
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=4, affinity='cosine', linkage='average')
predict = cluster.fit_predict(df_new[['xportAmt','mrktGrwth','imprtAmt']])
predict
```
